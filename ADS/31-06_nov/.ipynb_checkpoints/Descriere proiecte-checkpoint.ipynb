Conţinut

Seturile de date conțin tranzacții efectuate cu carduri de credit în septembrie 2013 de către deținătorii de carduri europene.
Acest set de date prezintă tranzacțiile care au avut loc în două zile, unde avem 492 de fraude din 284.807 de tranzacții. Setul de date este foarte dezechilibrat, clasa pozitivă (fraude) reprezintă 0,172% din toate tranzacțiile.
